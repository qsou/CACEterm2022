For this project, We are supposed to find the best car to sell to a customer based on many factors. First, I imported all of the necessary packages into this Jupyter notebook. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

Here is where I read in the data sets and stored them into pandas dataframes.

```python
cars1 = pd.read_csv('cars1.csv')
```

```python
cars1
```

```python
cars2 = pd.read_csv('cars2.csv')
```

```python
cars2
```

Here is where I remove all of the unnamed columns, which would allow all the columns from data one and two to match.

```python
cars1 = cars1.loc[:, ~cars1.columns.str.contains('^Unnamed')]
```

```python
cars1
```

Here is were I made the two dataframes into one.

```python
cars = pd.concat([cars1, cars2], axis=0)
```

```python
cars.info()
```

```python
cars
```

Here is where I drop all of the rows that contain all '?', which was necessary as it could not be turned into an integer.

```python
#has to drop all of the '?' to get the max horsepower
bestcars = cars[cars.horsepower != "?"]
```

```python
bestcars.info()
```

Here is where I changed the 'horsepower' column into an integer.

```python
#df["a"] = pd.to_numeric(df["a"])
#horsepower was an obj and needs to be changed into int to find max
bestcars["horsepower"] = pd.to_numeric(bestcars["horsepower"])
```

Here was where I looked at each column to see if there are any correlations between each other.

Here I took top 25 percent of the cars with the best mpg, since I am going toward an everyday approach. I wanted to recommend an everyday car, which mpg is our main concern as it depicts how much money we spend on gas. An everyday car to me is an car that is confortable to drive and has a good mpg, in addition with it being safe.

```python
fc = bestcars[bestcars.mpg > 29]
```

For the final dataset, I only placed cars with the top 25 percent of acceleration, and the reasoning behind this act is for safety. I have noticed that in an emergency if a car could get an extra boost quickly, you can avoid car crashes in most cases. There were many cases on the road that an extra boost helped people avoid car crashes.

```python
#dfObj[(dfObj['Sale'] > 30) & (dfObj['Sale'] < 33) ]
final = fc[fc['acceleration'] >= 18.1]
```

I found the car that weighs the lightest. Lighter cars go faster, but the downfall is that the faster you go the more likely the car will slip. This would require the driver to be more cautious and focus while driving this specific car.

```python
print(final[final.weight == final.weight.min()])
```
